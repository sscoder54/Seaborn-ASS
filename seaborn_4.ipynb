Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dimond = sns.load_dataset('diamonds')
```

```python
dimond.head()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x='price',data=dimond, hue='cut')
plt.xlabel("Price")
plt.ylabel("Count")
plt.title("Histogram of Diamond Prices by Cut")
```
