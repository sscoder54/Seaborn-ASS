Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn provides a variety of plots for visualizing different types of data.

Scatter Plot (sns.scatterplot()):
Use Case:
Scatter plots are useful for visualizing the relationship between two continuous variables. They help identify patterns, trends, and potential outliers in the data.

Bar Plot (sns.barplot()):
Use Case:
Bar plots are effective for comparing the mean or aggregate values of a categorical variable. They are commonly used to show the distribution of a variable across different categories.

Histogram (sns.histplot()):
Use Case:
Histograms are ideal for visualizing the distribution of a single variable. They display the frequency or count of observations within predefined bins, providing insights into the data's underlying structure.

Relplot (sns.relplot()):
Use Case:
A figure-level interface for drawing relational plots. It provides a high-level interface for creating scatter plots, line plots, and other types of plots that show the relationship between two variables.

sns.displot():
Use Case:
Used to display histograms, kernel density estimates (KDE), and rug plots to visualize the distribution of a single variable. Additionally, it supports faceting, allowing you to create subplots based on categorical variables.

Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.scatterplot(x ='timepoint',y = 'signal',data=fmri,hue='event')
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x ='timepoint',y = 'signal',data=fmri,hue='event',style='region')
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.title("FMRI Signal Over Time for Different Events and Regions")
```
